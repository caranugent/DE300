```python
# analyze the results: delay by state (mean)

# Filter the DataFrame for 'WeatherDelay' and 'OriginStateName' columns
selected_columns = df[['WeatherDelay', 'OriginStateName']]

# Group by 'OriginStateName' and calculate the mean weather delay for each state
mean_weather_delay_by_state = selected_columns.groupby('OriginStateName')['WeatherDelay'].mean()

# Plot the data
plt.figure(figsize=(12, 6))
mean_weather_delay_by_state.plot(kind='bar')
plt.title('Mean Weather Delay by State')
plt.xlabel('State')
plt.ylabel('Mean Weather Delay (minutes)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# analyze the results: total weather delays by state

# Filter the DataFrame for 'WeatherDelay' and 'OriginStateName' columns
selected_columns = df[['WeatherDelay', 'OriginStateName']]

# Group by 'OriginStateName' and calculate the mean weather delay for each state
mean_weather_delay_by_state = selected_columns.groupby('OriginStateName')['WeatherDelay'].sum()

# Plot the data
plt.figure(figsize=(12, 6))
mean_weather_delay_by_state.plot(kind='bar')
plt.title('Total Weather Delay by State')
plt.xlabel('State')
plt.ylabel('Total Weather Delay (minutes)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# analyze the results: cancellations by airline

# Filter the DataFrame for 'WeatherDelay' and 'OriginStateName' columns
selected_rows = df[['Reporting_Airline', 'Cancelled']]

# Group by 'OriginStateName' and calculate the mean weather delay for each state
mean_weather_delay_by_state = selected_rows.groupby('Reporting_Airline')['Cancelled'].size()

# Plot the data
plt.figure(figsize=(12, 6))
mean_weather_delay_by_state.plot(kind='bar')
plt.title('Total Number of Cancelled Flights by Airline')
plt.xlabel('Airline')
plt.ylabel('Total number of Cancelled Flights')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# analyze the results: percent cancellations by airline

# Group by 'Reporting_Airline' and calculate the total number of flights for each airline
total_flights_by_airline = df.groupby('Reporting_Airline').size()

# Group by 'Reporting_Airline' and 'Cancelled', then count the number of cancelled flights for each airline
cancelled_flights_by_airline = df[df['Cancelled'] == 1].groupby('Reporting_Airline').size()

# Calculate the ratio of cancelled flights to total flights for each airline
cancelled_ratio_by_airline = cancelled_flights_by_airline / total_flights_by_airline

# Plot the data
plt.figure(figsize=(12, 6))
cancelled_ratio_by_airline.plot(kind='bar')
plt.title('Percentage of Cancelled Flights by Airline')
plt.xlabel('Airline')
plt.ylabel('Percentage of Cancelled Flights')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
